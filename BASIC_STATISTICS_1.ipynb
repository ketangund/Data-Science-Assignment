# [Descriptive Analytics for Numerical Columns](https://)

```python
#Load the dataset:
import pandas as pd
```

```python

# 1.Load the dataset
df = pd.read_csv('/content/sales_data_with_discounts.csv')
```

```python
df
```

```python
df.describe()
```

```python
#2.Identify numerical columns:
numerical_columns = df.select_dtypes(include=['float64', 'int64']).columns
numerical_columns
```

```python
#3.Calculate basic statistical measures:
descriptive_stats = df[numerical_columns].describe().T
descriptive_stats['mean'] = df[numerical_columns].mean()
descriptive_stats['median'] = df[numerical_columns].median()
descriptive_stats['mode'] = df[numerical_columns].mode().iloc[0]
descriptive_stats['std'] = df[numerical_columns].std()
descriptive_stats
```

```python
#4.Interpretation:
#Mean: Provides the average value of the column.
#Median: The middle value when the data is ordered.
#Mode: The most frequently occurring value.
#Standard Deviation: Indicates the dispersion or spread of the data.
```

Data Visualization

Plot histograms:

```python
import matplotlib.pyplot as plt

df[numerical_columns].hist(bins=15, figsize=(15, 10))
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

Boxplots:


```python
df[numerical_columns].plot(kind='box', subplots=True, layout=(3, 3), figsize=(15, 10))
plt.suptitle('Boxplots of Numerical Columns')
plt.show()
```

Bar Chart Analysis for Categorical Columns:

```python
#Identify categorical columns:
categorical_columns = df.select_dtypes(include=['object']).columns
categorical_columns
```

```python
#Create bar charts:
for col in categorical_columns:
    df[col].value_counts().plot(kind='bar', figsize=(10, 5))
    plt.title(f'Bar Chart of {col}')
    plt.show()
```

# Standardization of Numerical Variables

```python
#Standardize numerical columns:
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_standardized = df.copy()
df_standardized[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Before and after comparison
df_before_after = pd.concat([df[numerical_columns], df_standardized[numerical_columns]], axis=1)
df_before_after.columns = [f'{col}_before' for col in numerical_columns] + [f'{col}_after' for col in numerical_columns]
df_before_after.head()
```

# Conversion of Categorical Data into Dummy Variables

```python
# Need for one-hot encoding:
#Machine learning algorithms require numerical input.
#One-hot encoding converts categorical variables into binary columns.
```

```python
#Apply one-hot encoding:
df_encoded = pd.get_dummies(df, columns=categorical_columns, drop_first=True)
df_encoded.head()


```
