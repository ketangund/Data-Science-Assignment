```python
# Handling missing values
# Assuming missing values are represented as '?'
df.replace('?', pd.NA, inplace=True)  # Replace '?' with NA
df.dropna(inplace=True)  # Drop rows with any missing values
```

```python
#Apply scaling techniques to numerical features:
#a. Standard Scaling:
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.preprocessing import StandardScaler

# Select numerical features for standard scaling
numerical_features = ['age', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']
```

```python
# Apply standard scaling
scaler = StandardScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
```

```python
#b. Min-Max Scaling:
from sklearn.preprocessing import MinMaxScaler
```

```python
# Apply Min-Max scaling
scaler = MinMaxScaler()
df[numerical_features] = scaler.fit_transform(df[numerical_features])
#Standard Scaling: Preferred when the distribution of the numerical feature is normal or close to normal. It centers the data around 0 with a standard deviation of 1, making it suitable for algorithms that assume zero-centered data.
#Min-Max Scaling: Preferred when the numerical feature does not follow a normal distribution and has outliers. It scales the data to a fixed range (usually between 0 and 1), preserving the shape of the original distribution.
```

# **3. Feature Engineering:**

```python
# Create new features
df['education_num_squared'] = df['education_num'] ** 2
df['capital_gain_log'] = np.log(df['capital_gain'] + 1)  # Adding 1 to handle zero values
```

```python
# Apply transformation to skewed numerical feature
skewed_feature = 'capital_gain_log'
df[skewed_feature] = np.log(df[skewed_feature] + 1)  # Log transformation to reduce skewness
```
